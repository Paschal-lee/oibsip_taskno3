```python
import pandas as pd
```

```python
de_videos = pd.read_csv("DEvideos.csv")
de_videos.head(10)
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process
print(de_videos.columns)

# Column renaming to proper
de_videos.columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
"Publish Time", "Tags", "Views", "Likes", "Dislikes", 
"Comment Count", "Thumbnail Link", "Comment Disabled",
"Ratings Disabled", "Video Error", "Description"]

de_videos.head(3)
```

```python
# Check for data types and correct any inconsistencies
de_videos.info()

# Convert "Trending Date" to datetime with a specific format
de_videos["Trending Date"] = pd.to_datetime(de_videos["Trending Date"], format="%y.%d.%m")

de_videos["Trending Date"] = pd.to_datetime(de_videos["Trending Date"])
de_videos["Publish Time"] = pd.to_datetime(de_videos["Publish Time"])
de_videos.head()
```

```python
# 2. Missing Data Handling

# Check for missing values
de_videos_missing = de_videos.isnull().sum()
de_videos_missing
```

```python
# Deleting rows with null values

de_videos.dropna(how = "all")
de_videos.dropna(subset = ["Description"], inplace = True)
de_videos_missing = de_videos.isnull().sum()
de_videos_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueeness

# Check for duplicate records
de_videos.duplicated().sum()
```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper 
de_videos.columns = de_videos.columns.str.title()
de_videos.head(10)
de_videos["Title"] = de_videos["Title"].str.title()
de_videos["Channel Title"] = de_videos["Channel Title"].str.title()
de_videos["Tags"] = de_videos["Tags"].str.title()
de_videos.head(10)
```

```python
# 5. Dropping Columns With unuseful dataset
columns_to_drop = ["Category Id", "Tags", "Thumbnail Link", "Description"]
de_videos.drop(columns=columns_to_drop, inplace=True)
de_videos
```
