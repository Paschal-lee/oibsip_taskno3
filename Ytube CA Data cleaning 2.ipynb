```python
import pandas as pd
```

```python
ca_videos = pd.read_csv("CAvideos.csv")
ca_videos.head(10)
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process
print(ca_videos.columns)

# Column renaming to proper
ca_videos.columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
"Publish Time", "Tags", "Views", "Likes", "Dislikes", 
"Comment Count", "Thumbnail Link", "Comment Disabled",
"Ratings Disabled", "Video Error", "Description"]

ca_videos.head(3)
```

```python
# Check for data types and correct any inconsistencies
ca_videos.info()

# Convert "Trending Date" to datetime with a specific format
ca_videos["Trending Date"] = pd.to_datetime(ca_videos["Trending Date"], format="%y.%d.%m")

ca_videos["Trending Date"] = pd.to_datetime(ca_videos["Trending Date"])
ca_videos["Publish Time"] = pd.to_datetime(ca_videos["Publish Time"])
ca_videos.head()
```

```python
# 2. Missing Data Handling

# Check for missing values
ca_videos_missing = ca_videos.isnull().sum()
ca_videos_missing
```

```python
# Deleting rows with null values

ca_videos.dropna(how = "all")
ca_videos.dropna(subset = ["Description"], inplace = True)
ca_videos_missing = ca_videos.isnull().sum()
ca_videos_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueeness

# Check for duplicate records
ca_videos.duplicated().sum()
```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper 
ca_videos.columns = ca_videos.columns.str.title()
ca_videos.head(10)
ca_videos["Title"] = ca_videos["Title"].str.title()
ca_videos["Channel Title"] = ca_videos["Channel Title"].str.title()
ca_videos["Tags"] = ca_videos["Tags"].str.title()
ca_videos.head(10)
```

```python
# 5. Dropping Columns With unuseful dataset
columns_to_drop = ["Category Id", "Tags", "Thumbnail Link", "Description"]
ca_videos.drop(columns=columns_to_drop, inplace=True)
ca_videos
```
