```python
import pandas as pd
```

```python
in_videos = pd.read_csv("INvideos.csv")
in_videos.head(10)
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process
print(in_videos.columns)

# Column renaming to proper
in_videos.columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
"Publish Time", "Tags", "Views", "Likes", "Dislikes", 
"Comment Count", "Thumbnail Link", "Comment Disabled",
"Ratings Disabled", "Video Error", "Description"]

in_videos.head(3)
```

```python
# Check for data types and correct any inconsistencies
in_videos.info()

# Convert "Trending Date" to datetime with a specific format
in_videos["Trending Date"] = pd.to_datetime(in_videos["Trending Date"], format="%y.%d.%m")

in_videos["Trending Date"] = pd.to_datetime(in_videos["Trending Date"])
in_videos["Publish Time"] = pd.to_datetime(in_videos["Publish Time"])
in_videos.head()
```

```python
# 2. Missing Data Handling

# Check for missing values
in_videos_missing = in_videos.isnull().sum()
in_videos_missing
```

```python
# Deleting rows with null values

in_videos.dropna(how = "all")
in_videos.dropna(subset = ["Description"], inplace = True)
in_videos_missing = in_videos.isnull().sum()
in_videos_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueness

# Check column names
print(in_videos.columns)

# List of columns for checking duplicates
duplicated_columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
                       "Publish Time", "Tags", "Views", "Likes", "Dislikes", 
                       "Comment Count", "Comment Disabled",
                       "Ratings Disabled", "Video Error"]

# Drop duplicates based on corrected column names
in_videos.drop_duplicates(subset=duplicated_columns, inplace=True)

# Check for any remaining duplicates
print(in_videos.duplicated().sum())
```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper 
in_videos.columns = in_videos.columns.str.title()
in_videos.head(10)
in_videos["Title"] = in_videos["Title"].str.title()
in_videos["Channel Title"] = in_videos["Channel Title"].str.title()
in_videos["Tags"] = in_videos["Tags"].str.title()
in_videos.head(10)
```

```python
# 5. Dropping Columns With unuseful dataset
columns_to_drop = ["Category Id", "Tags", "Thumbnail Link", "Description"]
in_videos.drop(columns=columns_to_drop, inplace=True)
in_videos
```
