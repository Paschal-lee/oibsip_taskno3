```python
import pandas as pd
```

```python
fr_videos = pd.read_csv("FRvideos.csv")
fr_videos.head(10)
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process
print(fr_videos.columns)

# Column renaming to proper
fr_videos.columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
"Publish Time", "Tags", "Views", "Likes", "Dislikes", 
"Comment Count", "Thumbnail Link", "Comment Disabled",
"Ratings Disabled", "Video Error", "Description"]

fr_videos.head(3)
```

```python
# Check for data types and correct any inconsistencies
fr_videos.info()

# Convert "Trending Date" to datetime with a specific format
fr_videos["Trending Date"] = pd.to_datetime(fr_videos["Trending Date"], format="%y.%d.%m")

fr_videos["Trending Date"] = pd.to_datetime(fr_videos["Trending Date"])
fr_videos["Publish Time"] = pd.to_datetime(fr_videos["Publish Time"])
fr_videos.head()
```

```python
# 2. Missing Data Handling

# Check for missing values
fr_videos_missing = fr_videos.isnull().sum()
fr_videos_missing
```

```python
print(fr_videos.columns)
```

```python
# Deleting rows with null values

fr_videos.dropna(how = "all")
fr_videos.dropna(subset = ["Description"], inplace = True)
fr_videos_missing = fr_videos.isnull().sum()
fr_videos_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueness

# Check column names
print(fr_videos.columns)

# List of columns for checking duplicates
duplicated_columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
                       "Publish Time", "Tags", "Views", "Likes", "Dislikes", 
                       "Comment Count", "Comment Disabled",
                       "Ratings Disabled", "Video Error"]

# Drop duplicates based on corrected column names
fr_videos.drop_duplicates(subset=duplicated_columns, inplace=True)

# Check for any remaining duplicates
print(fr_videos.duplicated().sum())

```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper
#fr_videos.columns = fr_videos.columns.str.title()
#fr_videos.head(10)
fr_videos["Title"] = fr_videos["Title"].str.title()
fr_videos["Channel Title"] = fr_videos["Channel Title"].str.title()
fr_videos["Tags"] = fr_videos["Tags"].str.title()
fr_videos.head(10)
```

```python
# 5. Dropping Columns With unuseful dataset
columns_to_drop = ["Category Id", "Tags", "Thumbnail Link", "Description"]
fr_videos.drop(columns=columns_to_drop, inplace=True)
fr_videos
```
