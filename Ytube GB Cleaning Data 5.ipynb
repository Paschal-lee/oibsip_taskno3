```python
import pandas as pd
```

```python
gb_videos = pd.read_csv("GBvideos.csv")
gb_videos.head(10)
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process
print(gb_videos.columns)

# Column renaming to proper
gb_videos.columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
"Publish Time", "Tags", "Views", "Likes", "Dislikes", 
"Comment Count", "Thumbnail Link", "Comment Disabled",
"Ratings Disabled", "Video Error", "Description"]

gb_videos.head(3)
```

```python
# Check for data types and correct any inconsistencies
gb_videos.info()

# Convert "Trending Date" to datetime with a specific format
gb_videos["Trending Date"] = pd.to_datetime(gb_videos["Trending Date"], format="%y.%d.%m")

gb_videos["Trending Date"] = pd.to_datetime(gb_videos["Trending Date"])
gb_videos["Publish Time"] = pd.to_datetime(gb_videos["Publish Time"])
gb_videos.head()
```

```python
# 2. Missing Data Handling

# Check for missing values
gb_videos_missing = gb_videos.isnull().sum()
gb_videos_missing
```

```python
# Deleting rows with null values

gb_videos.dropna(how = "all")
gb_videos.dropna(subset = ["Description"], inplace = True)
gb_videos_missing = gb_videos.isnull().sum()
gb_videos_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueness

# Check column names
print(gb_videos.columns)

# List of columns for checking duplicates
duplicated_columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
                       "Publish Time", "Tags", "Views", "Likes", "Dislikes", 
                       "Comment Count", "Comment Disabled",
                       "Ratings Disabled", "Video Error"]

# Drop duplicates based on corrected column names
gb_videos.drop_duplicates(subset=duplicated_columns, inplace=True)

# Check for any remaining duplicates
print(gb_videos.duplicated().sum())

```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper 
gb_videos.columns = gb_videos.columns.str.title()
gb_videos.head(10)
gb_videos["Title"] = gb_videos["Title"].str.title()
gb_videos["Channel Title"] = gb_videos["Channel Title"].str.title()
gb_videos["Tags"] = gb_videos["Tags"].str.title()
gb_videos.head(10)
```

```python
# 5. Dropping Columns With unuseful dataset
columns_to_drop = ["Category Id", "Tags", "Thumbnail Link", "Description"]
gb_videos.drop(columns=columns_to_drop, inplace=True)
gb_videos
```
