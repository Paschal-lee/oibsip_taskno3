```python
import pandas as pd
```

```python
us_videos = pd.read_csv("USvideos.csv")
us_videos.head(10)
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process
print(us_videos.columns)

# Column renaming to proper
us_videos.columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
"Publish Time", "Tags", "Views", "Likes", "Dislikes", 
"Comment Count", "Thumbnail Link", "Comment Disabled",
"Ratings Disabled", "Video Error", "Description"]

us_videos.head(3)
```

```python
# Check for data types and correct any inconsistencies
us_videos.info()

# Convert "Trending Date" to datetime with a specific format
us_videos["Trending Date"] = pd.to_datetime(us_videos["Trending Date"], format="%y.%d.%m")

us_videos["Trending Date"] = pd.to_datetime(us_videos["Trending Date"])
us_videos["Publish Time"] = pd.to_datetime(us_videos["Publish Time"])
us_videos.head()
```

```python
# 2. Missing Data Handling

# Check for missing values
us_videos_missing = us_videos.isnull().sum()
us_videos_missing
```

```python
# Deleting rows with null values

us_videos.dropna(how = "all")
us_videos.dropna(subset = ["Description"], inplace = True)
us_videos_missing = us_videos.isnull().sum()
us_videos_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueness

# Check column names
print(us_videos.columns)

# List of columns for checking duplicates
duplicated_columns = ["Video Id", "Trending Date", "Title", "Channel Title", "Category Id",
                       "Publish Time", "Tags", "Views", "Likes", "Dislikes", 
                       "Comment Count", "Comment Disabled",
                       "Ratings Disabled", "Video Error"]

# Drop duplicates based on corrected column names
us_videos.drop_duplicates(subset=duplicated_columns, inplace=True)

# Check for any remaining duplicates
print(us_videos.duplicated().sum())
```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper 
us_videos.columns = us_videos.columns.str.title()
us_videos.head(10)
us_videos["Title"] = us_videos["Title"].str.title()
us_videos["Channel Title"] = us_videos["Channel Title"].str.title()
us_videos["Tags"] = us_videos["Tags"].str.title()
us_videos.head(10)
```

```python
# 5. Dropping Columns With unuseful dataset
columns_to_drop = ["Category Id", "Tags", "Thumbnail Link", "Description"]
us_videos.drop(columns=columns_to_drop, inplace=True)
us_videos
```
