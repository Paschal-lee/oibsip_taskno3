```python
import pandas as pd
```

```python
airbnb = pd.read_csv("AB_NYC_2019.CSV")
airbnb
```

```python
# 1. Data Integrity: To ensure accuracy, consistency and reliality of data throughout the cleaning process

# Column renaming to proper
airbnb.columns = ["Id", "Name", "Host Id", "Host Name", "Neighbourhood Group",
"Neighbourhood", "Latitude", "Longitude", "Room Type", "Price", 
"Minimum Nights", "Number of Reviews", "Last Reviews",
"Monthly Reviews", "Calculated Host Listing Count", "Yearly Availability"]

airbnb.head(3)
```

```python
# Check for data types and correct any inconsistencies
airbnb.info()
airbnb["Last Reviews"] = pd.to_datetime(airbnb["Last Reviews"])
airbnb["Neighbourhood Group"] = airbnb["Neighbourhood Group"].astype("category")
airbnb["Room Type"] = airbnb["Room Type"].astype("category")
```

```python
# 2. Missing Data Handling

# Check for missing values
airbnb_missing = airbnb.isnull().sum()
airbnb_missing
```

```python
# Deleting rows with null values

airbnb.dropna(how = "all")
airbnb.dropna(subset = ["Name", "Host Name"], inplace = True)

# Impute missing values with mean for "Monthly Rewiews" and "Last rewiews" columns
airbnb["Last Reviews"].fillna(airbnb['Last Reviews'].mean(), inplace=True)
airbnb["Monthly Reviews"].fillna(airbnb['Monthly Reviews'].mean(), inplace=True)
```

```python
# Check for missing values
airbnb_missing = airbnb.isnull().sum()
airbnb_missing
```

```python
# 3. Duplicate Removal: Identifying and eliminating duplicate records to maintain data uniqueeness

# Check for duplicate records
airbnb.duplicated().sum()
```

```python
# 4. Standardization: To achieve consistent formatting across dataset

# Standardize column names to proper 
airbnb.columns = airbnb.columns.str.title()
airbnb.head(10)
airbnb["Name"] = airbnb["Name"].str.title()
airbnb["Host Name"] = airbnb["Host Name"].str.title()
airbnb["Neighbourhood Group"] = airbnb["Neighbourhood Group"].str.title()
airbnb["Neighbourhood"] = airbnb["Neighbourhood"].str.title()
airbnb["Room Type"] = airbnb["Room Type"].str.title()
airbnb.head(10)
```

```python
# 5. Outlier Detection: Identify and address outliers that may skew analysis or model performance.

# Detect outliers using Z-score 
airbnb_z_scores = (airbnb["Price"] - airbnb["Price"].mean()) / airbnb["Price"].std()
airbnb_outliers = airbnb[(airbnb_z_scores > 3) | (airbnb_z_scores < -3)]

# Remove outliers or apply transformations
airbnb = airbnb[(airbnb_z_scores < 3) & (airbnb_z_scores > -3)]
```
